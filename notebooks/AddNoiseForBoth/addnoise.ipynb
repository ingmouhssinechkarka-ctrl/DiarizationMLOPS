```python
#!/usr/bin/env python3
"""
Audio Dataset Mixer - Cr√©er 20 audios st√©r√©o en dialogue
Combine 1 audio du dataset Moroccan-Darija avec des segments du dataset DODa
"""

import os
import random
import numpy as np
import librosa
import soundfile as sf
from datasets import load_dataset
from pathlib import Path
import json

# Configuration
OUTPUT_DIR = "./dialogue_audios"
NUM_DIALOGUES = 20
SILENCE_DURATION = (0.5, 1.0)  # En secondes (al√©atoire entre min et max)

# D√©composition des segments DODa
SEGMENT_MAPPING = {
    "F1": (0, 999),
    "M3": (1000, 1999),
    "F2": (2000, 2730),
    "M1": (2731, 2800),
    "M2": (2801, 2999),
    "M2": (3000, 3999),
    "M1": (4000, 4999),
    "F3": (5000, 5999),
    "M1": (6000, 6999),
    "F4": (7000, 7999),
    "F1": (8000, 8999),
    "M2": (9000, 9999),
    "M1": (10000, 10999),
    "M1": (11000, 11999),
    "M2": (12000, 12350),
    "M1": (12351, 12742),
}

# Version corrig√©e avec indices uniques
SEGMENT_MAPPING = {
    "F1_0": (0, 999),
    "M3": (1000, 1999),
    "F2": (2000, 2730),
    "M1_0": (2731, 2800),
    "M2_0": (2801, 2999),
    "M2_1": (3000, 3999),
    "M1_1": (4000, 4999),
    "F3": (5000, 5999),
    "M1_2": (6000, 6999),
    "F4": (7000, 7999),
    "F1_1": (8000, 8999),
    "M2_2": (9000, 9999),
    "M1_3": (10000, 10999),
    "M1_4": (11000, 11999),
    "M2_3": (12000, 12350),
    "M1_5": (12351, 12742),
}

print("=" * 80)
print("üéôÔ∏è  AUDIO DATASET MIXER - Cr√©ation de dialogues st√©r√©o")
print("=" * 80)

# Cr√©er le dossier de sortie
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"\nüìÅ Dossier de sortie: {OUTPUT_DIR}")

# 1. Charger les datasets
print("\n‚è≥ Chargement des datasets...")
print("  - Moroccan-Darija-Wiki-Audio-Dataset")
try:
    dataset_moroccan = load_dataset("atlasia/Moroccan-Darija-Wiki-Audio-Dataset", split="train")
    print(f"    ‚úÖ {len(dataset_moroccan)} audios disponibles")
except Exception as e:
    print(f"    ‚ùå Erreur: {e}")
    exit()

print("  - DODa-audio-dataset")
try:
    dataset_doda = load_dataset("atlasia/DODa-audio-dataset", split="train")
    print(f"    ‚úÖ {len(dataset_doda)} audios disponibles")
except Exception as e:
    print(f"    ‚ùå Erreur: {e}")
    exit()

# 2. S√©lectionner 1 audio al√©atoire de Moroccan-Darija
print("\nüéµ S√©lection d'un audio source...")
moroccan_idx = random.randint(0, len(dataset_moroccan) - 1)
moroccan_audio = dataset_moroccan[moroccan_idx]

print(f"  ID: {moroccan_idx}")
print(f"  Titre: {moroccan_audio.get('title', 'N/A')}")

# Charger l'audio Moroccan
if isinstance(moroccan_audio['audio'], dict):
    audio_data = moroccan_audio['audio']['array']
    sr_moroccan = moroccan_audio['audio']['sampling_rate']
else:
    # Fallback si format diff√©rent
    audio_data = moroccan_audio['audio']
    sr_moroccan = 16000

print(f"  Dur√©e: {len(audio_data) / sr_moroccan:.2f}s @ {sr_moroccan}Hz")

# 3. D√©couper l'audio en segments selon la d√©composition
print("\n‚úÇÔ∏è  D√©coupage en segments...")
segments_data = []

for segment_name, (start_idx, end_idx) in SEGMENT_MAPPING.items():
    # Convertir indices en secondes (supposer 1 index = 1ms pour l'exemple)
    # Vous devez adapter selon votre vraie dur√©e
    start_time = start_idx / 1000.0  # en secondes
    end_time = end_idx / 1000.0

    # Convertir en samples
    start_sample = int(start_time * sr_moroccan)
    end_sample = int(end_time * sr_moroccan)

    # V√©rifier les limites
    if start_sample >= len(audio_data):
        print(f"  ‚ö†Ô∏è  {segment_name}: start > dur√©e audio, ignor√©")
        continue

    end_sample = min(end_sample, len(audio_data))

    if start_sample >= end_sample:
        print(f"  ‚ö†Ô∏è  {segment_name}: segment vide, ignor√©")
        continue

    segment_audio = audio_data[start_sample:end_sample]
    segment_duration = len(segment_audio) / sr_moroccan

    segments_data.append({
        'name': segment_name,
        'audio': segment_audio,
        'sr': sr_moroccan,
        'duration': segment_duration
    })

    print(f"  ‚úÖ {segment_name:<10}: {segment_duration:.3f}s")

print(f"\nüìä Total: {len(segments_data)} segments extraits")

# 4. G√©n√©rer 20 dialogues
print("\nüé¨ G√©n√©ration des 20 dialogues...")
metadata = []

for dialogue_num in range(NUM_DIALOGUES):
    print(f"\n  [{dialogue_num + 1}/{NUM_DIALOGUES}]", end=" ")

    left_channel = np.array([], dtype=np.float32)
    right_channel = np.array([], dtype=np.float32)

    try:
        # Pour chaque segment du Moroccan
        for segment in segments_data:
            segment_audio_left = segment['audio'].astype(np.float32) / 32768.0
            segment_duration = segment['duration']

            # S√©lectionner al√©atoirement un audio du segment DODa correspondant
            segment_name = segment['name']

            # Extraire le type de speaker (F1, M2, etc)
            speaker_type = segment_name.rsplit('_', 1)[0] if '_' in segment_name else segment_name

            # Trouver les audios correspondants dans DODa
            # (Ceci suppose que DODa a une colonne 'speaker' ou similaire)
            doda_candidates = [i for i in range(len(dataset_doda))
                               if speaker_type in str(dataset_doda[i].get('speaker', ''))]

            if not doda_candidates:
                print(f"‚ö†Ô∏è Pas d'audio DODa pour {speaker_type}, silence utilis√©")
                segment_audio_right = np.zeros_like(segment_audio_left)
            else:
                # S√©lectionner al√©atoirement
                doda_idx = random.choice(doda_candidates)
                doda_audio = dataset_doda[doda_idx]

                # Charger l'audio DODa
                if isinstance(doda_audio['audio'], dict):
                    doda_data = doda_audio['audio']['array']
                    sr_doda = doda_audio['audio']['sampling_rate']
                else:
                    doda_data = doda_audio['audio']
                    sr_doda = 16000

                # R√©√©chantillonner si n√©cessaire
                if sr_doda != sr_moroccan:
                    doda_data = librosa.resample(doda_data, orig_sr=sr_doda, target_sr=sr_moroccan)

                # Couper √† la m√™me dur√©e que le segment Moroccan
                doda_samples = int(segment_duration * sr_moroccan)
                segment_audio_right = doda_data[:doda_samples].astype(np.float32) / 32768.0

                # Padding si n√©cessaire
                if len(segment_audio_right) < len(segment_audio_left):
                    padding = np.zeros(len(segment_audio_left) - len(segment_audio_right), dtype=np.float32)
                    segment_audio_right = np.concatenate([segment_audio_right, padding])
                else:
                    segment_audio_right = segment_audio_right[:len(segment_audio_left)]

            # Ajouter les segments
            left_channel = np.concatenate([left_channel, segment_audio_left])
            right_channel = np.concatenate([right_channel, segment_audio_right])

            # Ajouter du silence entre les segments
            silence_duration = random.uniform(SILENCE_DURATION[0], SILENCE_DURATION[1])
            silence_samples = int(silence_duration * sr_moroccan)
            silence = np.zeros(silence_samples, dtype=np.float32)

            left_channel = np.concatenate([left_channel, silence])
            right_channel = np.concatenate([right_channel, silence])

        # Cr√©er l'audio st√©r√©o
        stereo_audio = np.stack([left_channel, right_channel], axis=1)

        # Normaliser si n√©cessaire
        max_val = np.max(np.abs(stereo_audio))
        if max_val > 1.0:
            stereo_audio = stereo_audio / max_val * 0.95

        # Sauvegarder
        output_file = os.path.join(OUTPUT_DIR, f"dialogue_{dialogue_num + 1:02d}.wav")
        sf.write(output_file, stereo_audio, sr_moroccan)

        duration_total = len(stereo_audio) / sr_moroccan
        print(f"‚úÖ {duration_total:.2f}s")

        # Sauvegarder m√©tadonn√©es
        metadata.append({
            'filename': f"dialogue_{dialogue_num + 1:02d}.wav",
            'moroccan_source_idx': moroccan_idx,
            'duration_seconds': duration_total,
            'sample_rate': sr_moroccan,
            'channels': 2
        })

    except Exception as e:
        print(f"‚ùå Erreur: {str(e)[:50]}")

# Sauvegarder m√©tadonn√©es
metadata_file = os.path.join(OUTPUT_DIR, "metadata.json")
with open(metadata_file, 'w') as f:
    json.dump(metadata, f, indent=2)

print("\n" + "=" * 80)
print(f"‚úÖ G√©n√©ration termin√©e!")
print(f"üìÅ {NUM_DIALOGUES} fichiers sauvegard√©s dans: {OUTPUT_DIR}")
print(f"üìã M√©tadonn√©es: {metadata_file}")
print("=" * 80)
```

```python
import os
import random
from pydub import AudioSegment

# --- CONFIGURATION ---
noise1_path = "noise1.wav"  # Chemin vers le bruit 1 (ex: centre d'appel calme)
noise2_path = "noise2.wav"  # Chemin vers le bruit 2 (ex: bruit de rue ou industriel)
input_folder = "/content/drive/MyDrive/GroundTruth_multispeaker_overlap/audioMono"        # Vos segments de 25s
output_folder = "/content/drive/MyDrive/GroundTruth_multispeaker_overlap_noise/audioMono" # Dossier de sortie

# Volume du bruit (ajustez si c'est trop fort ou trop faible)
noise_volume_adj = -10

# Cr√©er le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)

# --- FONCTION DE CHARGEMENT S√âCURIS√â (Fix frame alignment) ---
def load_and_fix_noise(path, volume_adj):
    print(f"Chargement de {os.path.basename(path)}...")
    audio = AudioSegment.from_file(path)

    # Correctif : Alignement des frames pour √©viter les erreurs pydub
    remainder = len(audio._data) % audio.frame_width
    if remainder > 0:
        print(f"üîß Correction de {remainder} octets orphelins sur {os.path.basename(path)}.")
        audio._data = audio._data[:-remainder]

    # Ajustement volume
    return audio + volume_adj

# 1. CHARGEMENT DES BRUITS
try:
    noise1_audio = load_and_fix_noise(noise1_path, noise_volume_adj)
    noise2_audio = load_and_fix_noise(noise2_path, noise_volume_adj)
except Exception as e:
    print(f"‚ùå Erreur lors du chargement des bruits : {e}")
    exit()

print("D√©but du traitement...")

# 2. LISTE DES FICHIERS ET DIVISION
files = sorted([f for f in os.listdir(input_folder) if f.endswith(('.wav', '.mp3'))])
total_files = len(files)
midpoint = total_files // 2  # Le point de bascule

print(f"Total fichiers : {total_files}")
print(f"files[0] √† files[{midpoint-1}] -> Bruit 1")
print(f"files[{midpoint}] √† files[{total_files-1}] -> Bruit 2")

# 3. BOUCLE DE TRAITEMENT
for i, filename in enumerate(files):
    original_path = os.path.join(input_folder, filename)

    try:
        # A. Charger le segment original
        segment = AudioSegment.from_file(original_path)
        segment_duration = len(segment) # Dur√©e en ms

        # B. Choisir quel bruit utiliser (50/50)
        if i < midpoint:
            current_noise = noise1_audio
            noise_name = "Bruit 1"
        else:
            current_noise = noise2_audio
            noise_name = "Bruit 2"

        # C. V√©rification de s√©curit√© (taille du bruit)
        if len(current_noise) < segment_duration:
            print(f"‚ö†Ô∏è Erreur : Le fichier {noise_name} est trop court pour {filename}")
            continue

        # D. Extraire un morceau de bruit al√©atoire
        # On prend un point de d√©part au hasard dans le fichier de bruit
        max_start = len(current_noise) - segment_duration
        start_pos = random.randint(0, max_start)

        noise_chunk = current_noise[start_pos : start_pos + segment_duration]

        # E. Superposer (Overlay)
        combined = segment.overlay(noise_chunk, position=0)

        # F. Exporter
        output_path = os.path.join(output_folder, filename)
        combined.export(output_path, format="wav")

        if i % 10 == 0: # Affiche un log tous les 10 fichiers pour ne pas spammer
            print(f"[{i+1}/{total_files}] Trait√© : {filename} avec {noise_name}")

    except Exception as e:
        print(f"‚ùå Erreur sur {filename}: {e}")

print("‚úÖ Termin√© ! Tous les fichiers ont √©t√© g√©n√©r√©s.")
```
