```python
import os
from pydub import AudioSegment

# --- CONFIGURATION ---
input_folder = "/content/drive/MyDrive/GroundTruth_multispeaker_overlap/audioStereo"
output_folder = "/content/drive/MyDrive/GroundTruth_multispeaker_overlap/audioMono"

# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)

print(f"Traitement des fichiers .wav dans : {input_folder}")

# Liste uniquement les fichiers .wav
wav_files = [f for f in os.listdir(input_folder) if f.lower().endswith('.wav')]

for i, filename in enumerate(wav_files):
    in_path = os.path.join(input_folder, filename)
    out_path = os.path.join(output_folder, filename)

    try:
        # 1. Charger
        audio = AudioSegment.from_file(in_path, format="wav")

        # 2. Convertir en Mono
        # set_channels(1) mixe les deux canaux (L+R) / 2
        audio = audio.set_channels(1)

        # 3. Sauvegarder (en forçant les paramètres standards : 16 bits PCM)
        # Vous pouvez ajouter bitrate="256k" si nécessaire, mais pour du wav c'est souvent inutile
        audio.export(out_path, format="wav", parameters=["-ac", "1"])

        print(f"[{i+1}/{len(wav_files)}] OK : {filename}")

    except Exception as e:
        print(f"ERREUR sur {filename} : {e}")
        # Si l'erreur "not a whole number of frames" revient ici,
        # on peut activer la méthode de secours :
        # audio = audio.split_to_mono()[0]
        # audio.export(out_path, format="wav")

print("--- Conversion terminée ---")
```
